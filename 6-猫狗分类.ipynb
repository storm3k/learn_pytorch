一个目录，如果包含`__init__.py` ，就成了一个包(package)

例如 data/ 文件夹下有`__init__.py`

`main.py` 可以用 `from data.dataset import DogCat`

如果 `__init__.py`中写入 `from .dataset import DogCat`

则`main.py`中可以直接写成`from data import DogCat`

或者 `import data`;`dataset = data.DogCat`

比`from data.dataset import DogCat`更便捷
